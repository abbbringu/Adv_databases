# Student

```python
import pandas as pd
import requests

df = pd.read_csv("./data/students.csv")
df = df[df["Student name"].str.startswith("Student ")]

PREFIXES = """PREFIX :   <http://www.semanticweb.org/user/ontologies/2025/3/untitled-ontology-3/>
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>
"""

triples = []
for _, row in df.iterrows():
    # student_uri     = f":student/{row['Student id']}"
    prog_iri= f"<http://www.semanticweb.org/user/ontologies/2025/3/untitled-ontology-3/{row['Programme']}>"
    subj = f":{row['Student id']}"
    subj = subj.replace(" ", "_")
    triples.append(f"{subj} rdf:type       :Student ;")
    triples.append(f"        :name         \"{row['Student name']}\" ;")
    triples.append(f"        :personId         \"{str(row['Student id'])}\" ;")
    triples.append(f"        :StudentInProgramme   {prog_iri} ;")
    triples.append(f"        :yearJoined  \"{row['Year']}\"^^xsd:integer ;")
    graduated = str(row['Graduated']).lower()
    triples.append(f"        :graduated    \"{graduated}\"^^xsd:boolean .")
    triples.append("")

body = "INSERT DATA {\n  " + "\n  ".join(triples) + "\n}"

sparql = PREFIXES + body


with open("insert_students.rq", "w") as f:
    f.write(sparql)
print("Wrote insert_students.rq")

endpoint = "http://localhost:7200/repositories/DIT930/statements"
headers  = {"Content-Type": "application/sparql-update"}

resp = requests.post(endpoint, data=sparql.encode("utf-8"), headers=headers)
if resp.ok:
    print("Successfully inserted all students")
else:
    print("SPARQL update failed:", resp.status_code, resp.text)

# test previuw
print(sparql[:500])
```

# Department

```python
import glob
import pandas as pd
import requests

# parse through all CSV files
csv_files = glob.glob("./data/*.csv")

#  unique department IDs from the two possible column names
dept_vals = set()
for path in csv_files:
    df = pd.read_csv(path, dtype=str)  # read everything as string
    for col in ("Department name", "Department"):
        if col in df.columns:
            dept_vals.update(df[col].dropna().unique())

print(f"Found {len(dept_vals)} unique departments:", dept_vals)

# 3) Build SPARQL INSERT DATA
PREFIXES = """PREFIX :   <http://www.semanticweb.org/user/ontologies/2025/3/untitled-ontology-3/>
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>

"""

triples = []
for dept in sorted(dept_vals):
    # make a safe fragment (no spaces) for the subject URI
    frag = dept.replace(" ", "_")
    subj = f"{frag}"

    triples.append(f":{subj} rdf:type           :Department ;")
    triples.append(f"    :departmentId    \"{dept}\"^^xsd:string .")
    triples.append("")  # blank line

body = "INSERT DATA {\n" + "\n".join("  " + line for line in triples) + "\n}"

sparql = PREFIXES + body

# 4a) Write out to inspect or upload manually
with open("insert_departments.rq", "w") as f:
    f.write(sparql)
print("Wrote insert_departments.rq")

# 4b) …or POST directly to GraphDB
endpoint = "http://localhost:7200/repositories/DIT930/statements"
headers  = {"Content-Type": "application/sparql-update"}
resp = requests.post(endpoint, data=sparql.encode("utf-8"), headers=headers)
if resp.ok:
    print("✅ Departments inserted successfully")
else:
    print("❌ SPARQL update failed:", resp.status_code, resp.text)
```

# Division

```python
import glob
import pandas as pd
import requests

# ——— 1) Find all CSV files ———
csv_files = glob.glob("./data/*.csv")

# ——— 2) Collect unique division codes ———
#    expecting values like "10054-01", "10011-02", etc.
div_codes = set()
for path in csv_files:
    df = pd.read_csv(path, dtype=str)
    if "Division name" in df.columns:
        div_codes.update(df["Division name"].dropna().unique())

print(f"Found {len(div_codes)} unique divisions:", div_codes)

# ——— 3) Build SPARQL INSERT DATA ———
PREFIXES = """PREFIX :   <http://www.semanticweb.org/user/ontologies/2025/3/untitled-ontology-3/>
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>

"""

triples = []
for code in sorted(div_codes):
    # code is "XX-YY"
    dept_id, div_id = code.split("-", 1)
    # make a URI‐safe fragment for the division, e.g. "10054-01" → "10054_01"
    frag = code
    subj = f":{frag}"
    dept_uri = f":{dept_id}"

    triples.append(f"{subj} rdf:type                :Division ;")
    triples.append(f"    :divisionId            \"{code}\" ;")
    triples.append(f"    :DivisionInDepartment  {dept_uri} .")
    triples.append("")  # blank line

body = "INSERT DATA {\n" + "\n".join("  " + line for line in triples) + "\n}"
sparql = PREFIXES + body

# ——— 4a) Write out to a .rq file ———
with open("insert_divisions.rq", "w") as f:
    f.write(sparql)
print("Wrote insert_divisions.rq")

# ——— 4b) Optionally POST directly to GraphDB ———
endpoint = "http://localhost:7200/repositories/DIT930/statements"
headers  = {"Content-Type": "application/sparql-update"}
resp = requests.post(endpoint, data=sparql.encode("utf-8"), headers=headers)
if resp.ok:
    print("✅ Divisions inserted successfully")
else:
    print("❌ SPARQL update failed:", resp.status_code, resp.text)

# Print the first 300 chars for a quick sanity check
print(sparql[:300])
```
